```python
#knn classifier

from sklearn.datasets import load_iris

iris=load_iris()

iris
```

```python
x=iris.data
y=iris.target

def train(x,y,size):
    idx=np.random.permutation(np.arange(x.shape[0]))
    ts=int(np.floor(size*x.shape[0]))
    testidx=idx[:ts]
    trainidx=idx[ts:]
    xtrain=x[trainidx,:]
    ytrain=y[trainidx]
    xtest=x[testidx,:]
    ytest=y[testidx]
    return xtrain,ytrain,xtest,ytest

```

```python
xtrain,ytrain,xtest,ytest=train(x,y,0.2)
```

```python
d=np.sum((xtrain[:,np.newaxis,:]-xtest[np.newaxis,:,:])*2,axis=2)
```
